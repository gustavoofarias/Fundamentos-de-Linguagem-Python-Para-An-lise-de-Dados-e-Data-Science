##Orientado a Objeto

```python
class Palavra:
    def __init__(self, palavra):
        self.palavra = palavra
        self.palavra_atual = ["_" for _ in palavra]

    def verificar_letra(self, letra):
        if letra in self.palavra:
            for idx, char in enumerate(self.palavra):
                if char == letra:
                    self.palavra_atual[idx] = letra
            return True
        return False

    def __str__(self):
        return " ".join(self.palavra_atual)

class Jogador:
    def __init__(self, nome, tentativas=6):
        self.nome = nome
        self.tentativas = tentativas

    def adivinhar_letra(self):
        letra = input("Adivinhe uma letra: ")
        return letra

class JogoDaForca:
    def __init__(self, palavra):
        self.palavra = Palavra(palavra)
        self.jogador = Jogador(nome="Jogador 1")

    def jogar(self):
        while self.jogador.tentativas > 0 and "_" in self.palavra.palavra_atual:
            print(self.palavra)
            letra = self.jogador.adivinhar_letra()
            if not self.palavra.verificar_letra(letra):
                self.jogador.tentativas -= 1
                print(f"Letra errada! Tentativas restantes: {self.jogador.tentativas}")

        if "_" not in self.palavra.palavra_atual:
            print(f"Parabéns {self.jogador.nome}, você ganhou!")
        else:
            print(f"Que pena {self.jogador.nome}, você perdeu! A palavra era {self.palavra.palavra}.")

# Exemplo de uso
jogo = JogoDaForca("python")
jogo.jogar()
s
```
