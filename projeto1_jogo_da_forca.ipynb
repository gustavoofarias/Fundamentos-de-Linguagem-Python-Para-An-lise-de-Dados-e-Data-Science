```python
palavra_secreta = "Gustavo"
tamanho_palavra = len(palavra_secreta)
letras_adivinhadas = [None for _ in range(tamanho_palavra)]
tentativas_restantes = 6
letras_erradas = []

def mostrar_palavra(letras_adivinhadas):
    for letra in letras_adivinhadas:
        if letra is None:
            print("_", end=" ")
        else:
            print(letra, end=" ")
    print()  # Corrigido: este print() deve estar fora do loop

while tentativas_restantes > 0 and None in letras_adivinhadas:
    mostrar_palavra(letras_adivinhadas)

    print(f"Faltam {tentativas_restantes} tentativas restantes.")
    if letras_erradas:
        print(f"Letras erradas: {letras_erradas}")

    letra = input("Adivinha uma letra:").upper()  # Corrigido: normaliza para maiúsculas para facilitar comparação

    if letra in palavra_secreta.upper():
        for i in range(tamanho_palavra):
            if palavra_secreta[i].upper() == letra:
                letras_adivinhadas[i] = palavra_secreta[i]  # Mantém a letra na forma original (maiúscula ou minúscula)
    else:
        if letra not in letras_erradas:
            letras_erradas.append(letra)  # Corrigido: append ao invés de apped
            tentativas_restantes -= 1
        else:
            print("Você já tentou essa letra.")

if None in letras_adivinhadas:
    print(f"Você perdeu! A palavra era: {palavra_secreta}")
else:
    print(f"Parabéns! Você adivinhou a palavra: {palavra_secreta}")


```
